```python
import pandas as pd
data = {
    'Name':['Rushab','Shailesh','Sumit'],
    'Age':[20,19,21],
    
    'city':['Pune','Mumbai','Delhi'],
    'TShirt_Size':['Large','Small','Medium'],
    'Number_Of_Pens':[0,5,1],
    'Test_Score':[83.9,97.0,78.5]
    }
df = pd.DataFrame(data)
print("Default DataTypes:")
print(df.dtypes)
size_order = ['Small','Medium','Large']
df['TShirt_Size'] = pd.Categorical(df['TShirt_Size'], categories = size_order, ordered = True)
print("\n COrrected DataTypes:")
print(df.dtypes)
```
